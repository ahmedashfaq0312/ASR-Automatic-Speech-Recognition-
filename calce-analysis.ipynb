# CALCE Dataset Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd

from calce.dataset import CALCEDataset
```

```python
# calce_root = "../RUL/datasets/CALCE/"
calce_root = "./CALCE/"
Battery_list = ['CS2_33', 'CS2_34', 'CS2_35', 'CS2_36', 'CS2_37', 'CS2_38']
```

```python
calce = CALCEDataset(batteries=Battery_list, calce_root=calce_root, clean_dataset=True)
```

```python
print(calce.data_dict[Battery_list[0]].head())
```

```python
def plot_curves(data_dict, plot_title):
    fig, ax = plt.subplots()
    for data_key in list(data_dict.keys()):
        ax.plot(calce.timestamps[data_key], data_dict[data_key], label=data_key)
    fig.legend()
    plt.title(plot_title)
```

```python
plot_curves(calce.capacities, "Capacities")
```

```python
plot_curves(calce.sohs, "State of Healths")
```

```python
plot_curves(calce.resistances, "Resistances")
```

```python
plot_curves(calce.ccct, "CCCTs")
```

```python
plot_curves(calce.cvct, "CVCTs")
```

```python
def get_and_plot_cycle_data(cycle_data, cycle):
    cycles = cycle_data[cycle_data["Cycle_Index"] == str(cycle)]
    time = cycles["Test_Time(s)"].astype(float)

    plt.figure()
    ax1 = plt.subplot2grid(shape=(2,6), loc=(0,0), colspan=2)
    ax1.plot(time, cycles["Current(A)"].astype(float))
    ax1.set_title("Current")
    ax2 = plt.subplot2grid((2,6), (0,2), colspan=2)
    ax2.plot(time, cycles["Voltage(V)"].astype(float))
    ax2.set_title("Voltage")
    ax3 = plt.subplot2grid((2,6), (0,4), colspan=2)
    ax3.plot(time, cycles["Internal_Resistance(Ohm)"].astype(float))
    ax3.set_title("Internal_Resistance")
    ax4 = plt.subplot2grid((2,6), (1,1), colspan=2)
    ax4.plot(time, cycles["Charge_Capacity(Ah)"].astype(float))
    ax4.set_title("Charge_Capacity")
    ax5 = plt.subplot2grid((2,6), (1,3), colspan=2)
    ax5.plot(time, cycles["Discharge_Capacity(Ah)"].astype(float))
    ax5.set_title("Discharge_Capacity")
    
    plt.tight_layout()
```

```python
for i in range(1, 11):
    get_and_plot_cycle_data(cycle_data, i)
```
