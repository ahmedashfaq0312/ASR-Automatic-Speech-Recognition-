# Dataset Cleanup
## NASA

```python
import matplotlib.pyplot as plt
from nasa.dataset import NASADataset
```

```python
def plot_data(data_dict, save_name):
    fig, ax = plt.subplots()
    for k, v in data_dict.items():
        ax.plot(v, label=k)
    ax.legend()
    plt.savefig(save_name)
```

```python
nasa_list = ["B0045", "B0046", "B0047", "B0048"]
```

```python
nasa_dataset_uncleaned = NASADataset(batteries=nasa_list, clean_dataset=False)
save_name = "nasa_" + "_".join(nasa_list) + "_uncleaned.png"
plot_data(nasa_dataset_uncleaned.capacities, save_name)
```

```python
nasa_dataset_cleaned = NASADataset(batteries=nasa_list)
save_name = "nasa_" + "_".join(nasa_list) + "_cleaned.png"
plot_data(nasa_dataset_cleaned.capacities, save_name)
```
