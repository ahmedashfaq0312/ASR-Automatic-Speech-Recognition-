```python
import pandas as pd

# Load the data
file_path = r"subset_file.csv"
data = pd.read_csv(file_path)

# Display the first few rows and data types
print(data.info())
print(data.head())
```

```python
# Group by timestamp and aggregate by mean
aggregated_data = data.groupby('Unnamed: 0').agg({
    'Step': 'first',  # Assuming 'Step' doesn't need averaging
    'Total_Time_Seconds': 'mean',
    'Voltage_V': 'mean',
    'Current_A': 'mean',
    'Charge_Capacity_Ah': 'mean',
    'Discharge_Capacity_Ah': 'mean',
    'Charge_Capacity_Wh': 'mean',
    'Discharge_Capacity_Wh': 'mean'
}).reset_index()
```

```python
# Replace zeros with NaN
aggregated_data.replace(0, pd.NA, inplace=True)

# Fill NaN values forward
aggregated_data.fillna(method='ffill', inplace=True)

# Alternatively, use interpolation
# aggregated_data.interpolate(method='linear', inplace=True)
```

```python
# Print the first few rows of the original and processed data
print("Original data:")
print(data.head(20))

print("Processed data:")
print(aggregated_data.head(20))
```

```python
import matplotlib.pyplot as plt

# Plot original data
plt.figure(figsize=(10, 5))
plt.plot(data['Unnamed: 0'], data['Voltage_V'], label='Original Voltage_V', color='blue')
plt.plot(data['Unnamed: 0'], data['Current_A'], label='Original Current_A', color='green')
plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.title('Original Data')
plt.legend()
plt.grid(True)
plt.show()

# Plot aggregated data
plt.figure(figsize=(10, 5))
plt.plot(aggregated_data['Unnamed: 0'], aggregated_data['Voltage_V'], label='Aggregated Voltage_V', color='red')
plt.plot(aggregated_data['Unnamed: 0'], aggregated_data['Current_A'], label='Aggregated Current_A', color='orange')
plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.title('Aggregated Data')
plt.legend()
plt.grid(True)
plt.show()
```
